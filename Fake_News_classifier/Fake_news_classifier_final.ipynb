```python
data.reset_index(inplace=True)
```

```python
def remove_punch(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

data['text'] = data['text'].apply(lambda x: remove_punch(x))
```

```python
data['text']
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

data['text'] = data['text'].apply(lambda x: tokenization(x.lower()))
```

```python
data['text']
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
data['text'] = data['text'].apply(lambda x: remove_stopwords(x))
```

```python
 data['text']
```

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

data['text'] = data['text'].apply(lambda x: stemming(x))
```

```python
data['text']
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

data['text'] = data['text'].apply(lambda x: lemmatizer(x))
```

```python
data['text']
```

```python
stemmer = nltk.stem.PorterStemmer()
data["text"]=data["text"].apply(lambda x:" ".join(stemmer.stem(token) for token in x))
```

```python
data["text"]
```

```python
from sklearn.model_selection import train_test_split
X, y = data['text'],data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators = 20).fit(X_train, y_train)
```

```python
RF_model.score(X_test, y_test)
```

```python
y_pred = RF_model.predict(X_test)
```

```python
CM = pd.crosstab(y_test, y_pred)
```
